# Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading the Iris dataset

```python
iris = sns.load_dataset('iris')
```

# dataset EDA 

```python
print("Shape of dataset:", iris.shape)
```

```python
print("\nColumn names:", iris.columns.tolist())
```

```python
print("\nFirst five rows:")
print(iris.head())
```

```python
print("\nDataset Info:")
print(iris.info())
```

```python
print("\nSummary Statistics:")
print(iris.describe())
```

# Data Visualization

# Scatter Plot (pairplot for relationships between features)

```python
sns.pairplot(iris, hue="species")
plt.suptitle("Scatter Plot of Iris Features", y=1.02)
plt.show()
```

# Histograms for distributions

```python
iris.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.show()
```

# Box Plots for outliers

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title("Box Plots of Iris Features")
plt.show()
```
